```python
X = train.drop("cc_cons", axis=1) # drop removes column so we removed it because that's what we want to predict

y = train["cc_cons"] # this stores the target variable separately
```

```python
X = X.drop("ID", axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,  # 80 % training data, 20 % testing data
    random_state=42 # to fix the random split
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# Import model to use 
from sklearn.linear_model import LinearRegression

# Creates an empty model
lr = LinearRegression()

# Train model, fit() means train 
lr.fit(X_train, y_train)

# Predict
y_pred = lr.predict(X_test)

# Evaluate the prediction error and how well the model explains the data
from sklearn.metrics import mean_squared_error, r2_score 

rmse = np.sqrt(mean_squared_error(y_test, y_pred)) # measures the average prediction error of the model
r2 = r2_score(y_test, y_pred) # value closer to 1 indicates better performance

print("RMSE:", rmse)
print("R2 Score:", r2)
```

```python
def rmspe(y_true, y_pred): # defining a new function
    return np.sqrt(np.mean(np.square((y_true - y_pred) / y_true))) * 100 

rmspe_value = rmspe(y_test, y_pred) # calls the function you created

print("RMSPE:", rmspe_value) # stores the result in rmspe_value
```

```python
# Import Model
from sklearn.tree import DecisionTreeRegressor 

# Create an empty Model
dt = DecisionTreeRegressor(random_state=42)

# Train the model
dt.fit(X_train, y_train)

# We ask it to predict the credit card consumption for customers in X_test
dt_pred = dt.predict(X_test)

# we compare the actual value and the predicted value, then calculate the average prediction error
dt_rmse = np.sqrt(mean_squared_error(y_test, dt_pred))

# it explain how much variation is present i.e., 66.5% in output
dt_r2 = r2_score(y_test, dt_pred)

# the prediction differs by 40% shown in output
dt_rmspe = rmspe(y_test, dt_pred)

print("Decision Tree RMSE:", dt_rmse)
print("Decision Tree R2:", dt_r2)
print("Decision Tree RMSPE:", dt_rmspe)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Create model, it creates 100 decision tree to predcit, 42 is a random split
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Train model
rf.fit(X_train, y_train)

# Predict
rf_pred = rf.predict(X_test)

# Compare the actual value with predicted value and then calculates the average error 
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))

# the variation is 83 % shown in output
rf_r2 = r2_score(y_test, rf_pred)

# means on average, the prediction differs by 28 %
rf_rmspe = rmspe(y_test, rf_pred)

print("Random Forest RMSE:", rf_rmse)
print("Random Forest R2:", rf_r2)
print("Random Forest RMSPE:", rf_rmspe)
```
